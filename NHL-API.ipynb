```python
import pandas as pd
```

```python
stats = pd.read_csv('stats2022NHL.csv')
bio = pd.read_csv('bioDataNHL.csv')
```

```python
stats.info()
```

```python
stats = stats.filter(items = ['id','stat.assists','stat.goals',
 'stat.shots','stat.hits','stat.penaltyMinutes','stat.shotPct','stat.plusMinus',
 'stat.points'])
```

```python
bio.info()
```

```python
bio.head()
```

```python
bio = bio.filter(items = ['id','fullName','primaryNumber','currentAge','birthCity','birthStateProvince','birthCountry',
                   'nationality','height','weight','shootsCatches','currentTeam.name','primaryPosition.abbreviation'])
```

```python
df = bio.merge(stats, on = 'id')
df = df[df['primaryPosition.abbreviation'] != 'G']
```

```python
top_nationalities = ['USA','SWE','CAN','FIN','RUS','CZE']
df = df[df['nationality'].isin(top_nationalities)]
```

```python
df.groupby(by = 'nationality')['stat.goals'].describe()
```

```python
df.groupby(by = 'nationality')['stat.assists'].describe()
```

```python
df.groupby(by = 'nationality')['stat.points'].describe()
```
